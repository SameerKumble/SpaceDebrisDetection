```python
import cv2
import numpy as np

def clear_debris(image_path):
    # Load the image
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Threshold the grayscale image to create a binary mask
    _, threshold = cv2.threshold(grayscale, 200, 255, cv2.THRESH_BINARY)

    # Finding contours in the binary mask
    contours, _ = cv2.findContours(threshold, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Filter the contours based on area and color intensity
    min_area = 1000  # Minimum contour area to consider
    min_intensity = 200  # Minimum intensity to consider as whitish
    whitish_contours = []

    for contour in contours:
        area = cv2.contourArea(contour)
        mean_intensity = cv2.mean(grayscale)[0]

        if area > min_area and mean_intensity > min_intensity:
            whitish_contours.append(contour)

    # Draw the contours on the original image with a red outline
    result = image.copy()
    cv2.drawContours(result, whitish_contours, -1, (0, 0, 255), 2)

    # Get the pixel addresses and values of the whitish regions
    pixel_info = []
    for contour in whitish_contours:
        for point in contour:
            x, y = point[0]
            pixel_value = grayscale[y, x]
            pixel_info.append(((x, y), pixel_value))

    # Display the result
    cv2.imshow("Result", result)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    # Return the pixel addresses with their values
    return pixel_info

# Example usage
image_path = r"C:\1111dip\1Subset\2.png"
detected_pixels = clear_debris(image_path)
print("Detected Pixel Addresses and Values:")
for pixel, value in detected_pixels:
    print(f"Pixel: {pixel}, Value: {value}")
```

```python
import cv2
import numpy as np

def debris(image_path):
    # Load the image
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Threshold the grayscale image to create a binary mask
    _, threshold = cv2.threshold(grayscale, 127, 255, cv2.THRESH_BINARY)

    # Invert the binary mask 
    inverted = cv2.bitwise_not(threshold)

    # Create a copy of the original image with swapped colors
    swapped_image = cv2.cvtColor(inverted, cv2.COLOR_GRAY2BGR)

    # Display the result
    cv2.imshow("Debris", swapped_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

# Example usage
image_path = r"C:\1111dip\1Subset\2.png"
debris(image_path)
```
